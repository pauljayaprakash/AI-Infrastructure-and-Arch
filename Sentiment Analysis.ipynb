```python
#IMporting the necessary libraries

import pandas as pd
from textblob import TextBlob
```

```python
#Loading the dataset
df = pd.read_csv("Airline_review.csv")

# Keeping only the Review column
df_clean = df[['Review']].copy()

# Droping missing values
df_clean = df_clean.dropna(subset=['Review'])

df_clean = df_clean.reset_index(drop=True)

df_clean.head()
```

```python
def get_polarity(text):
    return TextBlob(str(text)).sentiment.polarity

df_clean['polarity'] = df_clean['Review'].apply(get_polarity)
df_clean.head()
```

```python
def classify(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

df_clean['sentiment'] = df_clean['polarity'].apply(classify)
df_clean.head()
```

```python
df_clean.to_csv("sentiment_output.csv", index=False)
```
